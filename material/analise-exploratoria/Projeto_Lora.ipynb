```python
# Função para converter colunas para o tipo numérico após substituir vírgulas por pontos
def convert_column(column):
    if column.dtype == 'object':
        column = column.str.replace(',', '.')
    return pd.to_numeric(column, errors='ignore')

# Aplicar a função a todas as colunas do dataframe
basefinal_df = basefinal_df.apply(convert_column)
```

```python

import matplotlib.pyplot as plt
import numpy as np


# Filtrar apenas os dados de Pernambuco (PE) e São Paulo (SP)
estados_selecionados = ['PE', 'SP']

# Valores das médias
media_idh = [media_idh_por_uf[estado] for estado in estados_selecionados]
media_idh_edu = [media_idh_edu_por_uf[estado] for estado in estados_selecionados]
media_idh_long = [media_idh_long_por_uf[estado] for estado in estados_selecionados]
media_idh_renda = [media_idh_renda_por_uf[estado] for estado in estados_selecionados]

# Configuração do gráfico
x = np.arange(len(estados_selecionados))  # localização dos grupos no eixo X
largura = 0.2  # largura das barras

fig, ax = plt.subplots()

# Barras para cada tipo de média
barras_idh = ax.bar(x - 1.5*largura, media_idh, largura, label='IDH')
barras_edu = ax.bar(x - 0.5*largura, media_idh_edu, largura, label='IDH Educação')
barras_long = ax.bar(x + 0.5*largura, media_idh_long, largura, label='IDH Longevidade')
barras_renda = ax.bar(x + 1.5*largura, media_idh_renda, largura, label='IDH Renda')

# Adicionar os rótulos e o título
ax.set_xlabel('Estados')
ax.set_ylabel('Média')
ax.set_title('Médias de IDH e seus componentes para PE e SP')
ax.set_xticks(x)
ax.set_xticklabels(estados_selecionados)
ax.legend()

# Adicionar os valores acima das barras
def adicionar_valores(barras):
    for barra in barras:
        altura = barra.get_height()
        ax.annotate(f'{altura:.2f}',
                    xy=(barra.get_x() + barra.get_width() / 2, altura),
                    xytext=(0, 3),  # 3 pontos de deslocamento
                    textcoords="offset points",
                    ha='center', va='bottom')

adicionar_valores(barras_idh)
adicionar_valores(barras_edu)
adicionar_valores(barras_long)
adicionar_valores(barras_renda)

# Exibir o gráfico
plt.show()
```
